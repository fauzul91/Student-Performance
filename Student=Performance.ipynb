# Dataset Information

### Age

The age of the students ranges from 15 to 18 years

### Gender

Gender of the students, where 0 represents Male and 1 represents Female

### Ethnicity

Ethnicity: The ethnicity of the students, coded as follows:
0: Caucasian
1: African American
2: Asian
3: Other

### ParentalEducation

ParentalEducation: The education level of the parents, coded as follows:
0: None
1: High School
2: Some College
3: Bachelor's
4: Higher

### StudyHabits

Study Habits
StudyTimeWeekly: Weekly study time in hours, ranging from 0 to 20.
Absences: Number of absences during the school year, ranging from 0 to 30.
Tutoring: Tutoring status, where 0 indicates No and 1 indicates Yes.

### Parental Support

ParentalSupport: The level of parental support, coded as follows:
0: None
1: Low
2: Moderate
3: High
4: Very High

### Extracurriculer

Extracurricular: Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.

### Sports

Sports: Participation in sports, where 0 indicates No and 1 indicates Yes.

### Music

Music: Participation in music activities, where 0 indicates No and 1 indicates Yes.

### Volunteering

Volunteering: Participation in volunteering, where 0 indicates No and 1 indicates Yes.

### GPA

GPA: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities.

### Grade Class

GradeClass: Classification of students' grades based on GPA:
0: 'A' (GPA >= 3.5)
1: 'B' (3.0 <= GPA < 3.5)
2: 'C' (2.5 <= GPA < 3.0)
3: 'D' (2.0 <= GPA < 2.5)
4: 'F' (GPA < 2.0)

# Start Analysis

### Identifikasi Korelasi

```python
correlation_matrix = data.corr()

data.corr()

plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)

# Menambahkan judul
plt.title('Correlation Matrix')

# Menampilkan plot
plt.show()
```

### Pengaruh Parental Support Terhadap GPA

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix[['ParentalSupport']], annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation of Features with Parental Support')
plt.show()
```

Suatu korelasi dikatakan memiliki hubungan yang kuat jika hasilnya mendekati angka 1. Sebaliknya, suatu korelasi dikatakan memiliki hubungan yang lemah jika hasilnya mendekati angka 0. 

Berdasarkan matriks korelasi di atas, pengaruh antara ParentalSupport dengan GPA tidak terlalu kuat. Hasil yang didapatkan mendekati angka nol yaitu 0.19 sehingga tidak terlalu berpengaruh. Mari kita lihat korelasinya dalam bentuk diagram batang.

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='ParentalSupport', y='GPA', data=data)
plt.title('Pengaruh Parental Support Terhadap GPA Siswa')
plt.xlabel('Support Orang Tua')
plt.ylabel('GPA Siswa')
plt.xticks(rotation=45)
plt.show()
```

Meskipun korelasi antara ParentalSupport dengan GPA menunjukkan angka 0.19, bukan berarti ParentalSupport tidak memiliki pengaruh terhadap GPA siswa. Jika melihat grafik di atas, GPA siswa meningkat seiring dengan bertambahnya support dari orang tua.

### Absences With GPA

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix[['Absences']], annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation of Features with Absences')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(data['Absences'], data['GPA'], alpha=0.7, edgecolors='w', linewidths=0.5)
plt.title('Pengaruh Absences Terhadap GPA')
plt.xlabel('Absences')
plt.ylabel('GPA')
plt.show()
```

Melalui scatter plot di atas, kita dapat menyimpulkan bahwa semakin tinggi jumlah absen, GPA yang didapatkan  cenderung menurun. Hal ini sangat relevan dengan kehidupan pelajar. Ketika jumlah absen semasa sekolah tinggi, kemungkinan besar mereka akan tertinggal pelajaran sehingga berpengaruh terhadap GPA mereka.

### Study Time Weekly With GPA

```python
sns.histplot(data=data,x='StudyTimeWeekly',kde=True)
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix[['StudyTimeWeekly']], annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation of Features with StudyTimeWeekly')
plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.regplot(data=data, x='StudyTimeWeekly', y='GPA', scatter_kws={'s':50}, line_kws={'color': 'red'})

plt.title('Relationship between StudyTimeWeekly and GPA')
plt.xlabel('Study Time Weekly (hours)')
plt.ylabel('GPA')
plt.grid(True)
plt.show()
```

Berdasarkan matriks korelasi di atas, hasil StudyTimeWeekly dengan GPA siswa adalah 0.18. Data ini menunjukkan bahwa hubungan antara StudyTimeWeekly dengan GPA tergolong lemah karena mendekati angka nol. Jika melihat scatter plot di atas, GPA yang dihasilkan sangat beragam. Misal StudyTimeWeekly di 10 jam, terdapat GPA di angka 1.0, 0.5, 2.5, 3.75, dan seterusnya. Hasil ini menunjukkan bahwa GPA tidak dipengaruhi secara signifikan oleh StudyTimeWeekly
